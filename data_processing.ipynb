# Calculate the RUL

```python
# Making an array which contains EOL of all the Ids
EOL=[]
for sl_no in raw_data['ID']:
        EOL.append( ((raw_data[raw_data['ID'] == sl_no]["Cycle"]).values)[-1])
```

```python
# Append the "EOL" column
raw_data["EOL"]=EOL
# Calculate "RUL"
raw_data["RUL"] = raw_data["EOL"]-raw_data["Cycle"]
```
