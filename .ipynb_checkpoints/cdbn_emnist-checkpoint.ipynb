# Convolutional Deep Belief Networks on the EMNIST dataset


## Introduction

In this notebook I am going to explore and use an implementation of Convolutional Deep Belief Networks (CDBN): this type of model is obtained by stacking not normal Restricted Boltzmann Machines (RBMs), which have fully connected layers, but Convolutional RBMs (CRBMs), in which the layers are convolutional. This feature makes the model more scalable than normal DBNs for example in computer vision tasks, like the image classification task that I am going to perform. The paper I used as a reference is [Lee et al (2009)](https://web.eecs.umich.edu/~honglak/icml09-ConvolutionalDeepBeliefNetworks.pdf) which first introduced and tested the model.

## Implementation



The implementation of this model comes from this [github project](https://github.com/arthurmeyer/Convolutional_Deep_Belief_Network): I forked it and started working on it [in my own repository](https://github.com/SimoDR/Convolutional_Deep_Belief_Network). The model wasn't ready to use, mostly because of compatibility issues: the project was built on `tensorflow 0.12` and `python 3`. I brought it up to date with `python 3.7` and `tensorflow 2.7.0`. I birefly summarize the most relevant modifications I made:
*   replacing the legacy tensorflow functions with the newer ones, adapting the modified behaviour to the task they were originally intended for;
*   fixing a problem of NaN's values that would propagate during training: this was caused by the exponential activation function used in the *probability max pooling* layer which would lead to overflows. I solved this by replacing the exponential function with a similar, but less steep and more "bounded" *softplus* function; 
*   allowing the model to perform *semi-supervised* learning by accepting a dataset in which only a fraction of the data is provided with labels. This will be explored in the next sections;
*   implementing auto calculation of the input dimension for the deeper layers from the output dimension of the previous ones.

I will not dig into the details of this since it's not the main purpose of this report.

## Model



### Model building

The architecture of the model is well summarized in the output of the next cell. There are two **convolutional layers** (basically two CRBMs): each of them is composed by two sub-layers that are:
1.   convolutional filters (kernels): they perform the convolution on the image, becoming specialized in detecting specific features 
2.   probability max-pooling filters: they shrink the
representation of the convolutional layers by a constant
factor and compute the maximum activation (via exponential function or -- for this specific implementation -- softplus function) of the units in a small region of the convolutional layer.

The output of this becomes the input of the next layer.
These layers train with the *contrastive divergence* algorithm, thus making use of the "unsupervised" (unlabeled) training set.

On top of this architecture there is a **softmax layer** which is used for classification tasks: this is trained with the "supervised" (labeled) training set.

```python
# for the tensorflow graph session
tf.reset_default_graph()
sess = tf.Session()

batch_size = 200

my_cdbn = cdbn.CDBN(name = 'emnist_cdbn', 
                    batch_size = batch_size, 
                    path = '/home/', # saving the model
                    data = emnist_dataset, # data handler
                    session = sess, 
                    verbosity = 2)

# first layer
my_cdbn.add_layer('layer_1', fully_connected = False, 
                  v_height = 28, v_width = 28, v_channels = 1,    # input dimension
                  f_height = 7, f_width = 7, f_number = 30,       # convolutional filters dimension
                  init_biases_H = -5, init_biases_V = 0.01, init_weight_stddev = 0.01, 
                  gaussian_unit = True, gaussian_variance = 0.2,  # gaussian prior for the input layer
                  prob_maxpooling = True,                         # probability max pooling is used
                  padding = True,                                 # padding is used to preserve the original size of the input image
                  learning_rate = 0.0001, learning_rate_decay = 0.6, momentum = 0.5, decay_step = 50000, # learning rate parameters 
                  weight_decay = 1.0, sparsity_target = 0.01, sparsity_coef = 0.05) # sparsity and regularization parameters

# second layer
my_cdbn.add_layer('layer_2', fully_connected = False, 
                  v_height = "auto", v_width = "auto", v_channels = "auto", # input is automatically calculated from the previous layer output
                  f_height = 11, f_width = 11, f_number = 50, 
                  init_biases_H = -5, init_biases_V = 0.025, init_weight_stddev = 0.025, 
                  gaussian_unit = False, gaussian_variance = 0.2, 
                  prob_maxpooling = True, 
                  padding = True, 
                  learning_rate = 0.005, learning_rate_decay = 0.8, momentum = 0.9, decay_step = 50000,  
                  weight_decay = 0.1, sparsity_target = 0.1, sparsity_coef = 0.4)


# softmax layer
my_cdbn.add_softmax_layer(47,  # classes
                          0.1) # learning rate

my_cdbn.lock_cdbn()

# to be able to run this multiple times
tf.get_variable_scope().reuse_variables()
```

### Training

```python
# tf.debugging.enable_check_numerics() # to debug nan's

my_cdbn.manage_layers(['layer_1','layer_2'], # the layers to train
                      [],
                      [1500,3000], # for each layer, the number of training epochs
                      [2,2], # for each layer, length of the Gibbs chain in Gibbs sampling 
                      20000, # epochs for softmax training
                      restore_softmax = False, # for restoring a saved model
                      fine_tune = True)
```

The following function performs prediction on a given dataset on the trained CDBN model. I had to code it since I found no way to get it from the project implementation.

```python
def softmax_prediction(data,batches):

  data = np.reshape(data,(data.shape[0],28,28,1))

  if batches * batch_size > data.shape[0]:
    return -1

  # get the parameters of the softmax layer to perform predictions
  W = my_cdbn.W.eval(session=sess)
  b = my_cdbn.b.eval(session=sess)

  # the operation is batched because I get data from the tensorflow graph (which built the model
  # with a batch_size that regulates most of its internal operations). I found no
  # other way than to repeat this "fetching" operation with the size of batch_size
  for i in range(batches):
    batch_samples = tf.reshape(my_cdbn._get_input_level(2, data[i*batch_size:(i+1)*batch_size,:,:].astype('float32')).eval(session=sess), [batch_size, -1])
    # prediction
    y = tf.nn.softmax(tf.matmul(batch_samples, W) + b).eval(session=sess)
    predicted_class = tf.argmax(y,1).eval(session=sess)
    if i == 0:
      acc = predicted_class
    else:
      acc = np.hstack((acc,predicted_class))
  return acc
```

## Visualization

In this section I am going to visualize the internal representation of the model to gain a better understanding of it, and to show that it effectively builds a hierarchical representation of the input data.

### Filters
During the training of convolutional neural networks, matrices of weights are tuned. This matrices are called *filters* because, when applied to a sample input, they highlight some portion of the image that the network has learned to be relevant for the classification task. In the following sections I will plot the filters as 2d images. For the deeper layers (the second one) the visualization consists of multiple channel, because the filters of the deeper layers are applied on the output of each of the filters in the previous one.

```python
def plot_subplot(filters_layer,
                 num_channels = None, 
                 num_filters = None, 
                 labels = None,
                 labels_dict = None):
  
  if num_filters is None:
    num_filters = filters_layer.shape[3]
  if num_channels is None:
    num_channels = filters_layer.shape[2]
  for i in range(num_channels):
    print("--- Channel",i+1,"---")
    print_images(filters_layer[:,:,i],num_filters,labels,labels_dict) # makes use of print_images
```

#### Second filters as a linear combination of first layer

Filter for each base is computed as a weighted linear combination of the filters in the previous layer, where the weight to each filter in the previous layer is proportional to the sum of weights (weight vector of the network) originated from that base. This allows for a better visualization and is actually the kind of visualization that was used in the [Lee et al (2009)](http://robotics.stanford.edu/~ang/papers/icml09-ConvolutionalDeepBeliefNetworks.pdf) (figure 2 at page 5). The results don't really resemble the ones obtained in the paper, maybe a better hyperparameter configuration of the model would have resulted in a more similar visualization.

The function that I used is taken and adapted from another CDBN project, and the original version can be found [here](https://github.com/msolgi14/CDBN/blob/master/code/layer.py#L287).

```python
from scipy.signal import convolve 

def get_combination_deeper_filters(this_layer, prv_layer):

        filt_height = prv_layer.shape[0] + this_layer.shape[0] - 1
        filt_width = prv_layer.shape[1] + this_layer.shape[1] - 1

        filt_size = filt_height * filt_width
        all_filters = np.zeros((this_layer.shape[3], filt_size))
        
        for base_idx in range(this_layer.shape[3]):
            base_filter = np.zeros((filt_height, filt_width))
            for cnl_idx in range(prv_layer.shape[2]):
                reg = 1
                base_filter += convolve(this_layer[:,:,cnl_idx,base_idx]**reg, prv_layer[:,:,cnl_idx,0], 'full')

            all_filters[base_idx, :] = np.reshape(base_filter, filt_size)

        return all_filters
```

```python
# use the function
combination = get_combination_deeper_filters(filters_layer_2,filters_layer_1) 

# some reshaping
combination = combination.swapaxes(0,1)
combination = np.reshape(combination,(np.sqrt(combination.shape[0]).astype('int'),np.sqrt(combination.shape[0]).astype('int'),combination.shape[1]))

print_images(combination,combination.shape[2])
```

### Hidden layers

In this section there is a visualization of the *hidden layers* of each CRBM, computed on some given input images. These are also commonly known as *feature maps* and are useful to understand what each filter has been trained to detect. Note that:

1.   For the first hidden layer, the shape is `[14, 14, 30]`;
2.   for the first hidden layer, the shape is `[7, 7, 60]`

This is correct because after the convolution, the images are passed through the probability max pooling layer which shrink their sizes in half.

#### First hidden layer

This cell serves as a reference to the following ones by showing the original input images.

```python
print_images(supervised_training_images.swapaxes(0,1).reshape((28,28,supervised_training_images.shape[0])),
             num_subplots = 12,
             labels = data_train_labels,
             labels_dict = labels_dict)
```

```python
# Get the input from the bottom to the visible layer of the given level

def get_hidden_rappresentation(data,batches,n_layer):

  data = np.reshape(data,(data.shape[0],28,28,1))

  if batch_size * batches > data.shape[0]:
    return -1
    
  # I make use of a function of the model to get the hidden rappresentation, starting from
  # some input images. As before, since I rely on the internal configuration of the model, 
  # I have to repeat this computation in batches of the size of batch_size
  for i in range(batches):
    extracted_data = my_cdbn._get_input_level(n_layer, data[i*batch_size:(i+1)*batch_size,:,:].astype('float32')).eval(session=sess)
    if i == 0:
      hidden_layer = extracted_data
    else:
      hidden_layer = np.vstack((hidden_layer,extracted_data))
  return hidden_layer
```

It is clear how the first layer filters detect small portions of the images like edges. For example, the second filter (channel 2) seems to detect diagonal or vertical shapes, while the fifth (channel 5) is better activated by horizontal edges.

```python
first_hidden_layer = get_hidden_rappresentation(supervised_training_images,100,1)

# for each of the first 6 filters, I visualize the first 12 activations
plot_subplot(first_hidden_layer.swapaxes(0,1).swapaxes(1,2).swapaxes(2,3),
             num_channels = 6,
             num_filters = 12,
             labels=data_train_labels,
             labels_dict=labels_dict)
```

## Hierarchical representation

Deep Belief Networks are hierarchical generative model, as so it is possible to build a hierarchical rappresentation of the the input based on the learned weights. In the following cells, I will show how it is possible to cluster the classes of the dataset with the hidden rappresentation built by the two layers of the CDBN.  

```python
# we use this function to filter by class
def _get_mask(label,data):  
  return (data == label)

# here we filter by class and compute the mean
def _get_label_to_mean_hidd_repr(hidden_representation,data):
  return {
    label: hidden_representation[_get_mask(label,data)].mean(axis=0)  
    for label in range(47)
  }

# we use this to build the matrices
def _get_hidden_reprs_matrix(label_to_mean_hidd_repr):  
  return np.concatenate(
    [np.expand_dims(label_to_mean_hidd_repr[label], axis=0)  # here we adjust the shape of centroids to do the concat
    for label in range(47)])
```

With the help of the *hierarchical clustering* algorithm it is possible to visualize the hidden rappresentation as a special kind of graph called *dendogram*.

```python
import scipy.cluster as cluster

def _plot_dendrogram(mean_repr_matrix, title=""):
  fig, ax = plt.subplots(figsize=(20, 12))
  linkage = cluster.hierarchy.linkage(mean_repr_matrix, method="complete")  # we run the clustering algorithm here
  dendrogram = cluster.hierarchy.dendrogram(linkage,labels=list("0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZabdefghnqrt"))
  ax.set_title(title)
```

### First layer

To notice: to use hierarchical clustering, an extra step is needed if compared to regular DBNs. While with DBNs it is possible to reduce the hidden representation to the dimension of:

$$(c,n)$$

where $c$ = number of classes, and $n$ = neurons of the layer; in CDBNs every layer is composed by a number $n$ of $d$x$d$ matrices. Thus getting a representation of dimension:

$$(c,d,d,n)$$

To overcome this, two steps have been done:

1.   Reshaping the representation to $(c,d^2,n)$ by flattening the $d$x$d$ matrix;

2.   Taking the elementwise ("pixelwise") mean between the now flattened $n$ vectors (of dimention $d^2$) of each class, thus getting a final shape of:

$$(c,d^2)$$

Now it's possible to get a clustered representation with the hierarchical clustering algorithm.

```python
label_to_mean_hidd_repr_1 = _get_label_to_mean_hidd_repr(first_hidden_layer,data_train_labels[:first_hidden_layer.shape[0]])

mean_hidd_repr_matrix_1 = _get_hidden_reprs_matrix(label_to_mean_hidd_repr_1)

# 1.Reshape
mean_hidd_repr_matrix_1_rs = np.reshape(mean_hidd_repr_matrix_1,(mean_hidd_repr_matrix_1.shape[0],mean_hidd_repr_matrix_1.shape[1]**2,mean_hidd_repr_matrix_1.shape[3]))

# 2.Take the "elementwise" mean
mean_hidd_repr_matrix_1_rs_avg = mean_hidd_repr_matrix_1_rs.mean(axis = 2)
mean_hidd_repr_matrix_1_rs_avg.shape
```

The visualization shows that the model actually learned a hierarchical representation of the classes. The lower are the branches, the more similar are the samples of the classes with eachother. For example, the classes "L", "l" and "1" are clustered very closely to eachother. The same goes for "9", "g" and "q" which in fact look very similar.

```python
_plot_dendrogram(mean_hidd_repr_matrix_1_rs_avg, "First hidden layer")
```

### Second layer

I repeat the same process with the second layer.

```python
label_to_mean_hidd_repr_2 = _get_label_to_mean_hidd_repr(second_hidden_layer,data_train_labels[:first_hidden_layer.shape[0]])

mean_hidd_repr_matrix_2 = _get_hidden_reprs_matrix(label_to_mean_hidd_repr_2)

# 1.Reshape
mean_hidd_repr_matrix_2_rs = np.reshape(mean_hidd_repr_matrix_2,(mean_hidd_repr_matrix_2.shape[0],mean_hidd_repr_matrix_2.shape[1]**2,mean_hidd_repr_matrix_2.shape[3]))

# 2.Take the "elementwise" mean
mean_hidd_repr_matrix_2_rs_avg = mean_hidd_repr_matrix_2_rs.mean(axis = 2)
mean_hidd_repr_matrix_2_rs_avg.shape
```

For the second layer representation, the clustering algorithms detects fewer clusters (6 main clusters against the 8 of the first layer). This might suggest that the representation involves higher level features. For example, the black cluster seems to include the classes with a more overall "circular" shape (O,0,c,a); the classes in the light blue cluster seems to have a mix of round and sharp edges; the purple cluster seems to include shapes composed by mostly sharper edges. A better training configuration could lead to more precise results, but this hierarchical representation seems to, at least qualitatively, make sense.

```python
_plot_dendrogram(mean_hidd_repr_matrix_2_rs_avg, "Second hidden layer")
```

## Robustness to noise

In this sections I am going to compare the CDBN model with a standard Convolutional Neural Network (CNN) in various situations. The CNN has been chosen since it relies on the same convolutional approach of the CDBN: a comparison with a Feed Forward Neural Network would have been less fair.

The first comparison task is robustness to the injection of gaussian noise into the test images.

The CNN is built with two convolutional layers (each followed by a max pooling layer) and a fully connected one. It is trained on the same training set of the CDBN.

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, Conv2D, Flatten, MaxPooling2D
from tensorflow.python.framework.random_seed import set_random_seed

# this fuction trains the CNN on the training set and evaluates it on 
# the test set
def train_evaluate_cnn(
                   supervised_training_images,
                   supervised_training_labels,
                   test_images,
                   test_labels):

  np.random.seed(123)
  set_random_seed(2)

  cnn = Sequential()

  # first convolutional layer
  cnn.add(Conv2D(filters=128, kernel_size=(5,5), padding = 'same', 
                 activation='relu', input_shape=(28, 28,1)))
  cnn.add(MaxPooling2D(pool_size=(2,2), strides=(2,2)))

  # second convolutional layer
  cnn.add(Conv2D(filters=64, kernel_size=(3,3) , padding = 'same', 
                 activation='relu'))
  cnn.add(MaxPooling2D(pool_size=(2,2)))
  cnn.add(Flatten())

  # fully connected layer
  cnn.add(Dense(units=128, activation='relu'))

  # output layer
  cnn.add(Dense(units=47, activation='softmax'))

  cnn.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

  history = cnn.fit(supervised_training_images.reshape((supervised_training_images.shape[0],28,28,1)), 
                      supervised_training_labels, 
                      epochs=10, 
                      batch_size=200, 
                      verbose=1)
  

  accuracy_cnn_test = cnn.evaluate(test_images.reshape((test_images.shape[0],28,28,1)),
                           test_labels)[1]
                           
  return accuracy_cnn_test, cnn
```

```python
accuracy_cnn_test, cnn = train_evaluate_cnn(supervised_training_images,
                                            supervised_training_labels,
                                            test_images,
                                            test_labels)
```

The accuracy is a little higher than the CDBN (87.7% against 86%). We will see if this holds.

An example of an image injected with a 30% level of gaussian noise. It is still fairly recognizable as an "a".

```python
import torch
import torchvision as tv

# function to inject gaussian noise on a dataset. The noise level
# specifies the amount of noise that is added to each image.

def _inject_noise(data, noise_level):
  random_gaussian_tensor = torch.randn(data.shape)/6.6 + .6
  probability_tensor = torch.ones(data.shape)*noise_level
  mask_tensor = torch.bernoulli(probability_tensor)
  return data - mask_tensor * data + mask_tensor * random_gaussian_tensor


noise_level = 0.3
emnist_test_with_noise = _inject_noise(torch.tensor(supervised_training_images), noise_level)
__ = plt.imshow(emnist_test_with_noise[1].reshape(28, 28), cmap="gray")
```

```python
from sklearn.metrics import accuracy_score

# This function evaluates both models on the same noisy data

def _get_accuracy_values_at_noise_level(noise_level):
  
  # inject noise in data
  emnist_test_noisy = _inject_noise(torch.tensor(test_images), noise_level).numpy()[:2000]

  # evaluation of CDBN
  predicted_cdbn_noisy = softmax_prediction(emnist_test_noisy,batches=10)
  accuracy_cdbn = accuracy_score(predicted_cdbn_noisy,data_test_labels[:predicted_cdbn_noisy.shape[0]])

  # evaluation of CNN
  accuracy_cnn = cnn.evaluate(emnist_test_noisy.reshape((emnist_test_noisy.shape[0],28,28,1)),
                           test_labels[:emnist_test_noisy.shape[0]])[1]


  return accuracy_cdbn, accuracy_cnn
```

```python
# this function plots the results of the evaluation of both models
# on a grid of noise levels

def plot_noise_robustness_curves(noise_levels):
  accuracy_values_cdbn = []
  accuracy_values_cnn = []

  for noise_level in noise_levels:
    accuracy_cdbn, accuracy_cnn = _get_accuracy_values_at_noise_level(noise_level)
    accuracy_values_cdbn.append(accuracy_cdbn)
    accuracy_values_cnn.append(accuracy_cnn)

  fig, ax = plt.subplots()
  ax.plot(range(len(noise_levels)), accuracy_values_cdbn)
  ax.plot(range(len(noise_levels)), accuracy_values_cnn)

  ax.set_title("Robustness to noise")
  ax.set_xlabel("Noise level (%)")
  ax.set_ylabel("Accuracy")
  plt.xticks(range(len(noise_levels)), [int(l*100) for l in noise_levels])
  plt.legend(["CDBN", "CNN"])
```

The CDBN model shows a much better resistance to noise: the plot highlights how the performance deteriorates at a much lower rate than the CNN. Although starting a bit lower, the CDBN is able to hold and be 70% accurate at classifying images at a 30% level of noise corruption, while the CNN is less than 20% accurate at this point.

## Semi-supervised learning

One of the interesting features of DBNs is the ability to build a probablistic representation of data without needing labels for it, thus performing unsupervised learning. This last section will try to exploit this feature by performing semi-supervised learning: the CDBN model is trained on the whole EMNIST (balanced) training set but without labels, providing labeled examples only for a fraction of the training set, that are used to train the softmax layer. It will be interesting to see how it performs against a CNN trained only on the supervised fraction of the dataset.

```python
# this function wraps the training and evaluation of a CDBN. The model configuration
# is the same of the first section. The function will be used to train and evaluate 
# the model at different fractions of unsupervised and supervised training images.

def train_evaluate_cdbn(unsupervised_training_images,
                        supervised_training_images,
                        supervised_training_labels,
                        test_images,
                        test_labels):

  emnist_dataset = data_handler.DATA_HANDLER(unsupervised_training_images, 
                                            supervised_training_images, 
                                            supervised_training_labels, 
                                            test_images, 
                                            test_labels)

  tf.reset_default_graph()
  sess = tf.Session()

  batch_size = 200

  my_cdbn = cdbn.CDBN(name = 'emnist_cdbn', 
                      batch_size = batch_size, 
                      path = '/home/', 
                      data = emnist_dataset, 
                      session = sess, 
                      verbosity = 2)

  my_cdbn.add_layer('layer_1', fully_connected = False, v_height = 28, v_width = 28, v_channels = 1, 
                    f_height = 7, f_width = 7, f_number = 30, 
                    init_biases_H = -5, init_biases_V = 0.01, init_weight_stddev = 0.01, 
                    gaussian_unit = True, gaussian_variance = 0.2, 
                    prob_maxpooling = True, padding = True, 
                    learning_rate = 0.0001, learning_rate_decay = 0.6, momentum = 0.5, decay_step = 50000,  
                    weight_decay = 1.0, sparsity_target = 0.01, sparsity_coef = 0.05)

  my_cdbn.add_layer('layer_2', fully_connected = False, v_height = "auto", v_width = "auto", v_channels = "auto", 
                    f_height = 11, f_width = 11, f_number = 50, 
                    init_biases_H = -5, init_biases_V = 0.025, init_weight_stddev = 0.025, 
                    gaussian_unit = False, gaussian_variance = 0.2, 
                    prob_maxpooling = True, padding = True, 
                    learning_rate = 0.005, learning_rate_decay = 0.8, momentum = 0.9, decay_step = 50000,  
                    weight_decay = 0.1, sparsity_target = 0.1, sparsity_coef = 0.4)



  my_cdbn.add_softmax_layer(47, 0.1)

  my_cdbn.lock_cdbn()

  tf.get_variable_scope().reuse_variables()


  my_cdbn.manage_layers(['layer_1','layer_2'],[],[1500,3000], [2,2], 
                        int(supervised_training_images.shape[0]/5), 
                        restore_softmax = False, 
                        fine_tune = True)


  accuracy = my_cdbn.do_eval()

  sess.close()
  
  return accuracy[1]/100
```

```python
# utility function used to suppress the output of the training of the models when
# it is repeated multiple time

from contextlib import contextmanager,redirect_stderr,redirect_stdout
from os import devnull

@contextmanager
def suppress_stdout_stderr():
    """A context manager that redirects stdout and stderr to devnull"""
    with open(devnull, 'w') as fnull:
        with redirect_stderr(fnull) as err, redirect_stdout(fnull) as out:
            yield (err, out)
```

The supervised training set is built by taking a fraction of the whole dataset. In particular, to ensure that the new dataset is balanced, the same number of examples are taken from every class.

```python
from sklearn.utils import shuffle

# function used to extract the same number of examples from each of the 47
# classes. It is used to create reduced versions of the total dataset
# ensuring that they are still balanced.

def get_tot_examples_from_class(X,y,n_examples_for_class):
  positions = np.arange(X.shape[0])
  accumulator = []
  for i in range(47):
    # extract the positions of the samples of the i-th class, 
    position_masks = positions[y == i]
    # randomly sample only a part of them (n_examples_for_class)
    sampler = random.sample(range(len(position_masks)), n_examples_for_class)
    # accumulate them in a global vector that holds the sampled positions of
    # all the classes
    accumulator = np.hstack((accumulator,position_masks[sampler]))

  # take the actual samples from the positions
  X = np.take(X, accumulator.astype("int"), axis=0)
  y = np.take(y, accumulator.astype("int"))

  # shuffle the vectors in the same way
  X,y = shuffle(X,y)
  
  return X,y
```

```python
# this fuction trains and test both the CDBN and CNN models on a given fraction
# of labeled data per class

def test_models_with_supervised_fraction(supervised_training_images,
                                         supervised_training_labels,
                                         fractions_of_supervised_samples):
  cdbn_accuracy_accumulator = []
  cnn_accuracy_accumulator = []

  for fraction in fractions_of_supervised_samples:
    # calculate the number of example per class
    n = math.floor(supervised_training_images.shape[0]/47*fraction)

    print("For",fraction*100,"% of labeled data:",n,"examples per class")

    # vuild the dataset with n example per class
    supervised_training_images_sampled,supervised_training_labels_sampled = get_tot_examples_from_class(supervised_training_images,
                                                                                                        supervised_training_labels.argmax(1),
                                                                                                        n)
    supervised_training_labels_sampled = to_categorical(supervised_training_labels_sampled,47)

    print("- Training CDBN...", end=" ")
    # train and evaluate the CDBN
    with suppress_stdout_stderr(): # suppress training output
      cdbn_accuracy = train_evaluate_cdbn(unsupervised_training_images,
                                          supervised_training_images_sampled,
                                          supervised_training_labels_sampled,
                                          test_images,
                                          test_labels)
    cdbn_accuracy_accumulator.append(cdbn_accuracy)
    print('%0.02f accuracy' %(cdbn_accuracy))
    
    print("- Training CNN... ", end=" ")
    # train and evaluate the CNN
    with suppress_stdout_stderr(): # suppress training output
      cnn_accuracy,_ = train_evaluate_cnn(
                                          supervised_training_images_sampled,
                                          supervised_training_labels_sampled,
                                          test_images,
                                          test_labels)
    cnn_accuracy_accumulator.append(cnn_accuracy)
    print('%0.02f accuracy' %(cnn_accuracy))
    
  return cdbn_accuracy_accumulator, cnn_accuracy_accumulator
```

```python
# this function plot the results of the evaluation of the models on different 
# fractions of labeled data per class.

def plot_semi_supervised(fractions_of_supervised_samples):

  cdbn_accuracy_accumulator, cnn_accuracy_accumulator = test_models_with_supervised_fraction(supervised_training_images,
                                         supervised_training_labels,
                                         fractions_of_supervised_samples)

  fig, ax = plt.subplots()
  ax.plot(range(len(fractions_of_supervised_samples)), cdbn_accuracy_accumulator)
  ax.plot(range(len(fractions_of_supervised_samples)), cnn_accuracy_accumulator)

  ax.set_title("Accuracy at % of labeled data")
  ax.set_xlabel("Labeled data (%)")
  ax.set_ylabel("Accuracy")
  plt.xticks(range(len(fractions_of_supervised_samples)), [l*100 for l in fractions_of_supervised_samples])
  plt.legend(["CDBN", "CNN"])
```

As visible from the plot, the models are almost equivalent when provided with up to 5% of labeled data: they both score about 80% of accuracy on the test set. But below this threshold, the CDBN performs increasingly better than the CNN: for example, with 0.5% of labeled data (12 examples per class) the accuracy of the CDBN is 62% while the CNN is at 38%. With 4 examples per class: 41% accuracy (CDBN) agains 16% (CNN). Overall, the CDBN seems to benefit from the extra unlabeled data thanks to which it created a probabilistic representation. The output of the following cell provides more details into the accuracy scores at different levels of labeled data.
